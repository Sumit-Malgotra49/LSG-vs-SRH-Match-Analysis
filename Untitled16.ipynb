```python
import pandas as pd
import numpy as np
```

```python
url = '/content/match_data (2).csv'
df = pd.read_csv(url)
df.head(5)
```

```python
#checking the columns
df.columns
```

```python
conditions = [
    (df['Over Number'] <= 6),  # Powerplay: Over number is 6 or less
    (df['Over Number'] >= 7) & (df['Over Number'] <= 10),  # Upper Middle: Over number is between 7 and 10
    (df['Over Number'] >= 11) & (df['Over Number'] <= 15)  # Lower Middle: Over number is between 11 and 15
]

phase_names = ['Powerplay', 'Upper Middle', 'Lower Middle']

df['Phase'] = np.select(conditions, phase_names, default='Unknown')
df.head(5)
```

```python
srh_bowling = df[df['Bowling Team'] == 'Sunrisers Hyderabad']


srh_bowling['IsFour'] = srh_bowling['Batter Runs'] == 4
srh_bowling['IsSix'] = srh_bowling['Batter Runs'] == 6

# Group by 'Phase' and calculate metrics
grouped = srh_bowling.groupby('Phase').agg(
    Runs_Given=('Batter Runs', 'sum'),  # Total runs conceded
    Balls_Bowled=('Balls', 'count'),  # Total legal deliveries bowled
    Fours_Conceded=('IsFour', 'sum'),  # Total fours conceded
    Sixes_Conceded=('IsSix', 'sum')   # Total sixes conceded
).reset_index()

# Calculate Economy Rate (runs per over)
grouped['Economy_Rate'] = grouped['Runs_Given'] / (grouped['Balls_Bowled'] / 6)

# Calculate Total Boundaries (fours + sixes)
grouped['Boundaries'] = grouped['Fours_Conceded'] + grouped['Sixes_Conceded']
```

```python
grouped
```

```python
lsg_bowling = df[df['Bowling Team'] == 'Lucknow Super Giants']

lsg_bowling['IsFour'] = lsg_bowling['Batter Runs'] == 4
lsg_bowling['IsSix'] = lsg_bowling['Batter Runs'] == 6

# Group by 'Phase' and calculate metrics
grouped = lsg_bowling.groupby('Phase').agg(
    Runs_Given=('Batter Runs', 'sum'),  # Total runs conceded
    Balls_Bowled=('Balls', 'count'),  # Total legal deliveries bowled
    Fours_Conceded=('IsFour', 'sum'),  # Total fours conceded
    Sixes_Conceded=('IsSix', 'sum')   # Total sixes conceded
).reset_index()

# Calculate Economy Rate (runs per over)
grouped['Economy_Rate'] = grouped['Runs_Given'] / (grouped['Balls_Bowled'] / 6)

# Calculate Total Boundaries (fours + sixes)
grouped['Boundaries'] = grouped['Fours_Conceded'] + grouped['Sixes_Conceded']

grouped
```

 **SRH's bowling performance varies across phases, with the Powerplay seeing
most runs and boundaries, while the Lower Middle phase shows improved control. Strengthening the Upper Middle phase is key to reducing overall runs conceded.**

**LSG's bowling performance is most effective in the Upper Middle phase, where they concede fewer runs and maintain a strong economy rate. However, they face challenges in the Powerplay , leaking the most runs and boundaries, and struggle with consistency in the Lower Middle phase. Addressing these inconsistencies, particularly in high-pressure situations, will be key to improving their overall bowling effectiveness.**

```python
srh_total_runs = srh_batting['Runs'].sum()
srh_total_runs
```

```python
lsg_total_runs = lsg_batting['Runs'].sum()
lsg_total_runs
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: x = ['SRH', 'LSG'], y = [srh_total_runs, lsg_total_runs]
sns.barplot(x=['SRH', 'LSG'], y=[srh_total_runs, lsg_total_runs])

plt.title("Total Runs Comparison: SRH vs LSG")
plt.ylabel("Total Runs")
plt.xlabel("Team")
plt.show()
```
